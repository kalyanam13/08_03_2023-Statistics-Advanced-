### Q2. What are the types of Probability distribution?
### Answer : There are several types of probability distributions, and they can be broadly categorized into two main types: discrete probability distributions and continuous probability distributions.

### 1. Discrete Probability Distributions:

### Bernoulli Distribution: A distribution for a binary random variable that takes only two possible outcomes (success or failure), often used for modeling simple experiments.
### Binomial Distribution: Describes the number of successes in a fixed number of independent and identical Bernoulli trials.
### Poisson Distribution: Models the number of events occurring in a fixed interval of time or space, given the average number of occurrences.
### 2. Continuous Probability Distributions:

### Normal Distribution (Gaussian Distribution): Often referred to as the bell curve, it is characterized by its symmetrical shape and is widely used in various fields due to the central limit theorem.
### Uniform Distribution: All values in a given range are equally likely to occur.
### Exponential Distribution: Models the time between independent and identically distributed events occurring at a constant rate.
### Gamma Distribution: Generalizes the exponential distribution and is often used to model waiting times or the sum of certain types of random variables.
### Cauchy Distribution: Has heavy tails and no defined mean or variance, commonly used in physics and engineering.
### These distributions play a crucial role in probability theory and statistics, serving as mathematical models for various real-world phenomena. The choice of distribution depends on the nature of the random variable and the characteristics of the data being analyzed.

Here's a Python function that calculates the PDF of a normal distribution:

```python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution at a given point x.

    Parameters:
    - x: The point at which to evaluate the PDF.
    - mean: The mean of the normal distribution.
    - std_dev: The standard deviation of the normal distribution.

    Returns:
    The PDF value at the given point x.
    """
    coefficient = 1 / (math.sqrt(2 * math.pi) * std_dev)
    exponent_term = -(pow((x - mean), 2) / (2 * pow(std_dev, 2)))
    pdf_value = coefficient * math.exp(exponent_term)
    
    return pdf_value

# Example usage:
mean_value = 0  # Replace with your desired mean
std_dev_value = 1  # Replace with your desired standard deviation
point_to_evaluate = 1  # Replace with the point at which you want to calculate the PDF

pdf_result = normal_pdf(point_to_evaluate, mean_value, std_dev_value)
print(f"The PDF at x = {point_to_evaluate} is {pdf_result}")
```

### Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.
### Answer : Certainly! We can use the numpy library to generate a random sample from a binomial distribution and matplotlib for plotting the histogram. Here's a simple example in Python:

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the binomial distribution
n_trials = 1000
p_success = 0.4

# Generate a random sample from the binomial distribution
random_sample = np.random.binomial(n=1, p=p_success, size=n_trials)

# Plot the histogram
plt.hist(random_sample, bins=np.arange(-0.5, 2, 1), edgecolor='black', alpha=0.7)
plt.title('Binomial Distribution - Random Sample')
plt.xlabel('Number of Successes (0 or 1)')
plt.ylabel('Frequency')
plt.xticks([0, 1], ['Failure', 'Success'])
plt.show()
```

### This code uses numpy to generate a random sample of size 1000 from a binomial distribution with a probability of success (p) equal to 0.4. The plt.hist() function is then used to create a histogram of the results.

### Adjust the parameters as needed, and you can run this code in a Python environment to visualize the binomial distribution. The x-axis represents the number of successes (0 or 1), and the y-axis represents the frequency of each outcome in the random sample.

Here's a Python function that calculates the CDF of a Poisson distribution:

```python
import math

def poisson_cdf(k, mean):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution at a given point k.

    Parameters:
    - k: The point at which to evaluate the CDF.
    - mean: The mean of the Poisson distribution.

    Returns:
    The CDF value at the given point k.
    """
    cdf_value = 0
    for i in range(k + 1):
        cdf_value += math.exp(-mean) * (mean**i) / math.factorial(i)
    
    return cdf_value

# Example usage:
mean_value = 2.5  # Replace with your desired mean
point_to_evaluate = 3  # Replace with the point at which you want to calculate the CDF

cdf_result = poisson_cdf(point_to_evaluate, mean_value)
print(f"The CDF at k = {point_to_evaluate} is {cdf_result}")
```

### Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.
### Answer : Certainly! We can use the numpy library to generate a random sample from a Poisson distribution and then calculate the sample mean and variance. Here's an example in Python:

```python
import numpy as np

# Set the parameters for the Poisson distribution
mean_value = 5
sample_size = 1000

# Generate a random sample from the Poisson distribution
random_sample = np.random.poisson(lam=mean_value, size=sample_size)

# Calculate the sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample, ddof=1)  # ddof=1 for unbiased variance

# Print the results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

### This code uses numpy to generate a random sample of size 1000 from a Poisson distribution with a mean (λ) of 5. It then calculates the sample mean and variance using the np.mean() and np.var() functions, respectively.

### Feel free to modify the values of mean_value and sample_size as needed for your specific scenario.

### Q10. In normal distribution with respect to mean position, where does the least frequent data appear?
### Answer : In a normal distribution, also known as a Gaussian distribution or bell curve, the data is symmetrically distributed around the mean. The least frequent data appears in the tails of the distribution, farthest from the mean.

### Specifically:

### The least frequent data appears in the tails towards the extremes of both the left (negative side) and right (positive side) of the mean.
### The values in the tails become progressively less frequent as you move away from the mean in either direction.
### In a standard normal distribution (with a mean of 0 and a standard deviation of 1), about 68% of the data falls within one standard deviation of the mean, about 95% within two standard deviations, and about 99.7% within three standard deviations. This leaves only a small percentage of data in the tails beyond three standard deviations, making values in those regions less frequent.

### In summary, the least frequent data in a normal distribution appears in the tails, farthest from the mean.
